````python
import pandas as pd
from tabulate import tabulate


# Étape 1: Afficher le nombre de valeurs manquantes pour chaque attribut avant le traitement
missing_values_before = data.isnull().sum()

# Proposition et application d'une technique de traitement des valeurs manquantes
# Remplacer les valeurs manquantes par la médiane pour les colonnes numériques
numeric_columns = data.select_dtypes(include='number').columns
for column in numeric_columns:
    median_value = data[column].median()
    data[column] = data[column].fillna(median_value)

# Remplacer les valeurs manquantes par le mode pour les colonnes catégorielles
categorical_columns = data.select_dtypes(include='object').columns
for column in categorical_columns:
    mode_value = data[column].mode()[0]
    data[column] = data[column].fillna(mode_value)

# Étape 3: Vérifier de nouveau le nombre de valeurs manquantes
missing_values_after = data.isnull().sum()

# Préparer les données pour le tableau
table_data = [("Attribute", "Missing Values Before", "Missing Values After")]
attributes = data.columns
for attr in attributes:
    table_data.append((attr, missing_values_before[attr], missing_values_after[attr]))

# Utiliser tabulate pour afficher le tableau en format 'grid'
print(tabulate(table_data, headers="firstrow", tablefmt='grid'))
````

````python

# Supprimer la colonne Timestamp car elle ne sera pas utile pour la classification
data = data.drop(columns=['Timestamp'])

# Encoder les variables catégorielles en numériques
label_encoders = {}
for column in data.columns:
    if data[column].dtype == 'object':
        le = LabelEncoder()
        data[column] = le.fit_transform(data[column])
        label_encoders[column] = le


# Séparer les caractéristiques (features) et la cible (target)
X = data.drop(columns=['Mood_Swings'])
y = data['Mood_Swings']

# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Normaliser les données
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
````

###  **`k-NN`** :
n_neighbors (nombre des plus proches voisins : tester différentes
valeurs) et metric (mesure de distance : ‘euclidien’, ‘Manhattan’) ;
- Le modèle k-NN (K-Nearest Neighbors) est un algorithme de classification qui attribue une étiquette à une observation en fonction des étiquettes des exemples les plus proches dans l'espace des caractéristiques.

````python
# Définir le modèle k-NN et les hyperparamètres à tester
knn = KNeighborsClassifier()
knn_params = {
    'n_neighbors': [3, 5, 7, 9],
    'metric': ['euclidean', 'manhattan']
}

# Optimisation des hyperparamètres avec GridSearchCV
#cv=5 spécifie que la validation croisée en 5 folds sera utilisée pour évaluer chaque combinaison d'hyperparamètres.
start_time = time.time()
grid_knn = GridSearchCV(knn, knn_params, cv=5, scoring='accuracy', n_jobs=-1)
grid_knn.fit(X_train, y_train)
best_knn = grid_knn.best_estimator_

print(f"Best parameters for k-NN: {grid_knn.best_params_}")
````

- **Configuration de GridSearchCV**:

## Arbre de Decision:
Les arbres de décision sont des algorithmes de classification qui divisent les données en sous-groupes basés sur des règles de décision apprises à partir des caractéristiques des données. Ils sont faciles à interpréter et peuvent capturer des relations non linéaires entre les caractéristiques.

### Optimisation des Hyperparamètres pour un Arbre de Décision
#### Configuration des Hyperparamètres
- **Hyperparamètres Testés**:
  - `criterion`: Les critères pour mesurer la qualité d'une scission. Deux critères ont été testés:
    - `entropy` — mesure de l'impureté qui vise à maximiser l'information gagnée par chaque scission.
    - `gini` — mesure de l'impureté utilisée pour minimiser la probabilité de classification erronée.
#### Utilisation de GridSearchCV
- **Objectif**: Trouver le meilleur critère pour scinder les données en utilisant un arbre de décision.
